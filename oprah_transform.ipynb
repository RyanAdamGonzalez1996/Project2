```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
```

```python
url = 'https://en.m.wikipedia.org/wiki/Oprah%27s_Book_Club'
```

```python
tables = pd.read_html(url)
tables
```

```python
df = tables[0]
#df.columns = ['Date', 'Title', 'Author']
df
```

```python
new_df = df.drop(columns=['Reference'])

new_df
```

```python
url_2 = 'https://en.m.wikipedia.org/wiki/Oprah%27s_Book_Club_2.0'
```

```python
tables = pd.read_html(url_2)
tables
```

```python
df2 = tables[1]
df2
```

```python
new_df2 = df2.drop(columns=['Citation'])
new_df2
```

```python
#merged_df = pd.merge()

frames = [new_df, new_df2]

oprah = pd.concat(frames)

oprah = oprah.drop(columns=['Author'])



oprah
```

```python
oprah['Year'] = pd.DatetimeIndex(oprah['Date']).year
oprah['Month'] = pd.DatetimeIndex(oprah['Date']).month
oprah = oprah.drop(columns=['Date'])


oprah.head()
```

```python
#oprah.to_dict()

oprah_dict = oprah.set_index('Title').T.to_dict('list')

oprah_dict
```

```python
######read good books csv and add dictionaries to it

file_path = "Resources/good_books.csv"

df = pd.read_csv(file_path)
df
```

```python
#df[(df == pd.Series(oprah_dict)).all(1)]


df["Oprah Book Club (year, month)"] = df["title"].map(oprah_dict).fillna("N/A")

df
```

```python
df.to_csv("Output/oprah.csv", index=False, header=True)
```
